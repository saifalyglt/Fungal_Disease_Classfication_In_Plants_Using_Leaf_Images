# Step 2: Load Data And Apply Preprocessing And Augmentation

```python
# import ImageDataGenerator from tenserflow.keras.preprocesing.image for Data Preprocessing
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import pandas as pd
```

```python
# define Data augmentation and preprocessing
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2  # 20% of the data will be used for validation
)
```

```python
# Load  data
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),  # VGG16 input size
    batch_size=32,
    class_mode='categorical',
    shuffle=True,
    seed=42
)
```

```python
# Load training data
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=True,
    seed=42,
    subset='training'  # specify this is for training data
)
```

```python
# Load Validation Data
val_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False,
    seed=42,
    subset='validation'  # specify this is for validation data
)
```

```python
# import necessary Libraries for Manual Data Splitting
import os
import shutil
```

# Step 4: Load VGG16 pre-trained model


```python
# import necessary Libraries to Load VGG16 pre-trained model
from tensorflow.keras.applications import VGG16
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout
```

```python
# Load the VGG16 model without the top (fully connected) layers
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
# Create a new model with custom top layers
model = Sequential([
    base_model,
    Flatten(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(22, activation='softmax')
])
```

```python
# Freeze the pre-trained layers
for layer in base_model.layers:
    layer.trainable = False
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

# Step 5: Train the model

```python
# import laibary for TensorBoard Callback
from tensorflow.keras.callbacks import TensorBoard
```

```python
# Define TensorBoard callback
tensorboard = TensorBoard(log_dir='./logs')

from tensorflow.keras.callbacks import TensorBoard, ModelCheckpoint

# Define ModelCheckpoint callback
checkpoint = ModelCheckpoint('/content/gdrive/MyDrive/Capstone Project/best_model.h5',
                             save_best_only=True,
                             monitor='val_accuracy',
                             mode='max',
                             verbose=1)
```

```python
# Train the model using the generators
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10,
    callbacks=[tensorboard, checkpoint]
)
```

```python
# Save the model in h5 format
model.save('/content/gdrive/MyDrive/Capstone Project/model(ef=10).h5')
```

```python
# Function to preprocess the user-provided image and make a prediction
def predict_image(user_image_path, model):
    # Load and preprocess the user-provided image
    img = image.load_img(user_image_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array = img_array / 255.0  # Normalize pixel values to [0, 1]

    # Make a prediction using the loaded model
    prediction = model.predict(img_array)
    predicted_class_index = np.argmax(prediction)

    # Get class labels from the generator or define them based on your classes
    class_labels = list(train_generator.class_indices.keys())

    # Display the image and the predicted class
    plt.imshow(img)
    plt.title(f'Predicted Class: {class_labels[predicted_class_index]}')
    plt.axis('off')
    plt.show()

```
